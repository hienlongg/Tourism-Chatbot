```python
import urllib.request
from PIL import Image
from io import BytesIO

def read_image_from_url(url):
    """
    Reads an image from a given URL and returns a PIL Image object.
    """
    try:
        with urllib.request.urlopen(url) as url_response:
            image_data = url_response.read()
            img = Image.open(BytesIO(image_data))
            return img
    except Exception as e:
        print(f"Error reading image from URL: {e}")
        return None

# Example usage:
image_url = "http://localhost:5173/api/upload/image/image_20251205_103338_4462ae11.jpg"  # Replace with an actual image URL
image = read_image_from_url(image_url)

if image:
    image.show()  # Display the image
    # You can also perform other operations on 'image' here, e.g., save it, resize it, etc.
```
