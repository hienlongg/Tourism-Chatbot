```python
# keep rows where TenDiaDanh, DiaChi and ImageURL are non-null and not empty/whitespace
cols = ['TenDiaDanh', 'DiaChi', 'ImageURL']
filtered_data = data.dropna(subset=cols).copy()
for c in cols:
    filtered_data = filtered_data[filtered_data[c].astype(str).str.strip() != '']

filtered_data = filtered_data.reset_index(drop=True)
```

```python
import re
from pathlib import Path
from urllib.parse import urlparse, parse_qs

def make_direct_url(url):
    if not isinstance(url, str) or not url.strip():
        return ''
    url = url.strip()
    # Google Drive: /file/d/ID or ?id=ID
    m = re.search(r'drive\.google\.com\/file\/d\/([^\/\?]+)', url)
    if m:
        fid = m.group(1)
        return f'https://drive.google.com/uc?export=download&id={fid}'
    if 'drive.google.com' in url:
        q = parse_qs(urlparse(url).query)
        if 'id' in q:
            return f'https://drive.google.com/uc?export=download&id={q["id"][0]}'
    # Dropbox: turn dl=0 into dl=1 (direct download)
    if 'dropbox.com' in url:
        if 'dl=0' in url:
            return url.replace('dl=0', 'dl=1')
        if 'dl=1' in url:
            return url
        return url + '?dl=1'
    # Imgur: convert image page to direct image when possible
    parsed = urlparse(url)
    if parsed.netloc.endswith('imgur.com') and not re.search(r'\.(jpg|jpeg|png|gif)$', parsed.path, re.I):
        if parsed.path.startswith(('/a/', '/gallery/')):
            return url  # album/gallery - keep as-is
        return url.rstrip('/') + '.jpg'
    return url

# Use existing filtered_data DataFrame from earlier cells
out_df = filtered_data[['TenDiaDanh', 'ImageURL']].copy()
out_df['DirectImageURL'] = out_df['ImageURL'].astype(str).apply(make_direct_url)
out_df = out_df[['TenDiaDanh', 'DirectImageURL']]

out_dir = Path('../data/processed')
out_dir.mkdir(parents=True, exist_ok=True)
out_path = out_dir / 'danh_sach_dia_danh_images.csv'
out_df.to_csv(out_path, index=False, encoding='utf-8')
print('Saved:', out_path)
```

```python
import unicodedata
import re

def slugify(value):
    """
    Chuyển đổi chuỗi (có dấu, cách, ký tự đặc biệt) 
    thành một chuỗi an toàn để làm tên file/thư mục.
    Ví dụ: "Khu nhà công tử Bạc Liêu" -> "khu_nha_cong_tu_bac_lieu"
    """
    # Chuyển 'đ' thành 'd'
    value = str(value).replace("đ", "d").replace("Đ", "D")
    # Bỏ dấu tiếng Việt
    value = unicodedata.normalize('NFKD', value).encode('ascii', 'ignore').decode('utf-8')
    # Xóa ký tự đặc biệt, giữ lại chữ, số, khoảng trắng, gạch nối
    value = re.sub(r'[^\w\s-]', '', value).strip().lower()
    # Thay thế khoảng trắng/gạch nối bằng 1 gạch dưới
    value = re.sub(r'[\s-]+', '_', value)
    return value
```
